```python
#Disponibilité de matière grasse en quantité (g/personne/jour)
#Disponibilité de protéines en quantité (g/personne/jour)
#a passer en (Kg/personne/jour) => /1000
da['Disponibilité de matière grasse en quantité (g/personne/jour)']=da['Disponibilité de matière grasse en quantité (g/personne/jour)']/1000
da['Disponibilité de protéines en quantité (g/personne/jour)']=da['Disponibilité de protéines en quantité (g/personne/jour)']/1000
```

```python
#renomer les colonnes 
da = da.rename(columns={"Zone": "Pays"},inplace = False)
```

######  Fichier Sous Nutrition

```python
#ouvir le fichier de l'insecurité alimentaire( sous nutrition)
sn = pd.read_csv('sous_nutrition.csv')
```

### 1. Proportion de personnes en état de sous-nutrition

```python
#Ne retenir que l'année 2017
q1_2017=q1.loc[q1['Année'] == 2017, :]
```

### 2. Nombre théorique de personnes qui pourraient être nourries
     (à partir de la disponibilité alimentaire mondiale) 

Equation

●	La disponibilité intérieure (DI) correspond à la disponibilité totale d’un aliment au sein du pays.

Disponibilité intérieure = Nourriture + Pertes + Semences + Traitement
Disponibilité intérieure = Production + Importations - Exportations + Variation de stock

Donc

Production + Importations - Exportations +/- Variation de stock = DI = Nourriture + Pertes + Semences + Traitement


```python
# Fusion des dataset dispo_alimentaire et population pour l'année 2017 sur la colonne Pays
q2=da.merge(pop.loc[pop['Année'] == 2017, ['Pays','Pop. en Millions d’hab.']],on='Pays')
```

```python
#Vérification de la fusion
q2.head()
```

```python
#Calcul de la Disponibilité alimentaire monde
da_monde=((q2['Disponibilité alimentaire (Kcal/personne/jour)']*q2['Pop. en Millions d’hab.']).sum())*365
print("La disponibilité alimentaire mondiale annuelle et de" ,da_monde,"Kcal")
```

```python
# Calcul du nombre de personnes pouvant etre nourrie en prennant un besoin journalier de 2400kcal
print("Le nombre de personne pouvant être nourrie est de",round(da_monde/(2400*365)))
print("ce qui represente",round(((da_monde/(2400*365))/(q2['Pop. en Millions d’hab.']).unique().sum())*100,2),"%","de la population mondiale 2017")
```

### 3. Nombre théorique de personnes qui pourraient être nourries avec des aliments d'origines vegetales
     (à partir de la disponibilité alimentaire mondiale) 

```python
# Fusion des dataset dispo_alimentaire et population pour l'année 2017 sur la colonne Pays
# Selection de l'origine des aliments = vegetale
q3=da.merge(pop.loc[pop['Année'] == 2017, ['Pays','Pop. en Millions d’hab.']],on='Pays')
q3=q3[(q3['Origine'] == 'vegetale')]
q3.head()
```

```python
# Calcul du nombre de personnes pouvant etre nourrie en prennant un besoin journalier de 2400kcal
print("Le nombre de personne pouvant être nourrie avec des produits végétaux est de",round(da_monde_vege/(2400*365)))
print("ce qui represente",round(((da_monde_vege/(2400*365))/(q3['Pop. en Millions d’hab.']).unique().sum())*100,2),"%","de la population mondiale 2017")
```
